```python
import csv
import matplotlib.pyplot as pp
import os

logs_dir = input("Insert logs directory: ")

fig_index = 0

for log in os.scandir(logs_dir):
    if log.is_file():
        with open(log.path) as logfile:
            cases = logfile.read().strip().rstrip("---")
            cases = cases.split("---")

            indices = []
            makespans = []
            index = 0
            for case in cases:
                data = csv.reader(case.split('\n'))
                for row in data:
                    if not row == []:
                        indices.append(int(row[0]))
                        makespans.append(int(row[1]))
                plt = pp.figure(fig_index, figsize=(16, 9))
                pp.suptitle(f"Case {index}")
                pp.title(f"{log.name}")
                pp.xlabel("Iterations")
                pp.ylabel("Makespan")
                pp.plot(indices, makespans)
                plt.show()
                indices.clear()
                makespans.clear()
                index += 1
                fig_index += 1
```
